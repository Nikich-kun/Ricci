```python
import os
import cv2
import numpy as np
import argparse
import csv
import shutil
import sys
sys.path.append("..")
from Ricci import *
import matplotlib
from PIL import Image
import matplotlib.pyplot as plt

###########_ Normal img with Geometry weight #####################

DER  = '/home/alexander/Desktop/med/train/dataset/train/'


img_n = cv2.imread(str(DER +'n_b4_223_362_.png'))

R_ng = Ricci_img_g(img_n, k_size = 1)

width_ng, height_ng = R_ng.shape
R_new_ng = np.reshape(R_ng,(width_ng,height_ng))

plt.figure(figsize=(20,10))
plt.subplot(2,2,1),plt.title('Org'),plt.imshow(img_n)
plt.subplot(2,2,2),plt.title('Ricci for Normal with Geometry weight '),plt.imshow(R_ng)
plt.subplot(2,2,3),plt.title('Hist for Normal with Geometry weight'),plt.hist(R_new_ng, 200)
plt.show()

###########_ Tumor img with Geometry weight #####################

img_t = cv2.imread(str(DER +'t_b4_251_215_.png'))

R_tg = Ricci_img_g(img_t, k_size = 1)

width_tg, height_tg = R_tg.shape
R_new_tg = np.reshape(R_tg,(width_tg,height_tg))

plt.figure(figsize=(20,10))
plt.subplot(2,2,1),plt.title('Org'),plt.imshow(img_t)
plt.subplot(2,2,2),plt.title('Ricci for Tumor with Geometry weight'),plt.imshow(R_tg)
plt.subplot(2,2,3),plt.title('Hist for Tormal with Geometry weight'),plt.hist(R_new_tg)
plt.show()

###########_ Normal img with Combinatorial weight #####################

R_nc = Ricci_img_c(img_n)

width_nc, height_nc = R_nc.shape
R_new_nc = np.reshape(R_nc,(width_nc,height_nc))

plt.figure(figsize=(20,10))
plt.subplot(2,2,1),plt.title('Org'),plt.imshow(img_n)
plt.subplot(2,2,2),plt.title('Ricci for Normal with Combinatorial weight'),plt.imshow(R_nc)
plt.subplot(2,2,3),plt.title('Hist for Normal with Combinatorial weight'),plt.hist(R_new_nc)
plt.show()

###########_ Tumor img with Combinatorial weight #####################

R_tc = Ricci_img_c(img_t)

width_tc, height_tc = R_tc.shape
R_new_tc = np.reshape(R_tc,(width_tc,height_tc))

plt.figure(figsize=(20,10))
plt.subplot(2,2,1),plt.title('Org'),plt.imshow(img_t)
plt.subplot(2,2,2),plt.title('Ricci for Tumor with Combinatorial weight'),plt.imshow(R_tc)
plt.subplot(2,2,3),plt.title('Hist for Tumor with Combinatorial weight'),plt.hist(R_new_tc)
plt.show()
```
